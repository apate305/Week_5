```python
#Make a function that can take any non-negative integer 
#as an argument and return it with its digits in descending order. 
#Essentially, rearrange the digits to create the highest possible number. 
#If nonumber is passed in, make the default value inputted = 111111.

#Input = 43905 
#Output = 95430 
#Input = 584366 
#Output = 866543
```

```python
def highest_possible_number(n=111111):
    str_n = str(n)
    a_lst = []
    final_n = ""
    for i in str_n:
        a_lst.append(int(i))
    
    a_lst.sort(reverse=True)
    
    
    for i in a_lst:
        final_n += str(i)
    return final_n
```

```python
highest_possible_number(4583)
```

```python
import pandas as pd
import numpy as np

#STEP 1: Import jeopardy.csv

file1 = "JEOPARDY_CSV copy.csv"

# Read File
read_file = pd.read_csv(file1)
read_file.head()
```

```python
#STEP 2: Storing into/creating Pandas data frame from data

jeopardy_df = pd.DataFrame(read_file)
jeopardy_df.head()
```

```python
#a. How many rows are there in the dataframe? What does a row represent
#All the shows that were aired (not unique)
rows_count = len(jeopardy_df.index)
rows_count
```

```python
#b. How many unique shows were there each year?
    #There were 3640 unique shows that aired all together as seen below.
unique_jeop_count = jeopardy_df.nunique()
unique_jeop_count
```

```python
#b. How many unique shows were there each year? (cont)
    #jeapordy dataframe is grouped by date and count of each column is taken. The numbers in the "show number"
    #column indicate the number of shows that aired on the given date. 
grouped_year_df = jeopardy_df.groupby(' Air Date')
count_year_df = grouped_year_df.count()
count_year_df
```

```python
#b. How many unique shows were there each year? (cont)
    #All together 216930 shows were aired.
sum_shows = count_year_df["Show Number"].sum() 
sum_shows
```

```python
#b. How many unique shows were there each year? (cont)
    #dropping duplicates
temp_jeop_df = jeopardy_df
#b. How many unique shows were there each year? (cont)
    #1 unique show was aired each year. 
jeop_simplified_df = temp_jeop_df.drop_duplicates(subset = ['Show Number'])
grouped_simplified_years_df = jeop_simplified_df.groupby(' Air Date')
grouped_simplified_years_df.count()
```
